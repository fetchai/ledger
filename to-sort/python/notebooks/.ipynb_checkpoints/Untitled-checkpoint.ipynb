```python
def grads(X, Y, weights):
    grads = np.empty_like(weights)
    a = feed_forward(X, weights)
    delta = a[-1] - Y # cross-entropy
    grads[-1] = np.dot(a[-2].T, delta)
    for i in xrange(len(a)-2, 0, -1):
        delta = np.dot(delta, weights[i].T) * d_sigmoid(a[i])
        grads[i-1] = np.dot(a[i-1].T, delta)
    return grads / len(X)
```

```python
sigmoid = lambda x: 1 / (1 + np.exp(-x))
d_sigmoid = lambda y: y * (1 - y)

trX, trY, teX, teY = mnist.load_data(one_hot=True)

weights = [
    np.random.randn(28*28, 100) / np.sqrt(28*28),
    np.random.randn(100, 10) / np.sqrt(100)]
num_epochs, batch_size, learn_rate = 30, 10, 0.2

## Micro program
for i in xrange(num_epochs): 

    ## Running empoc
    for j in xrange(0, len(trX), batch_size):
        X, Y = trX[j:j+batch_size], trY[j:j+batch_size]
        weights -= learn_rate * grads(X, Y, weights)

    ## Evaluating prediction
    prediction = np.argmax(feed_forward(teX, weights)[-1], axis=1)
    print i, np.mean(prediction == np.argmax(teY, axis=1))
```

```python
import timeit
def testGen(n):
    A = libfetchcore.fetch.math.linalg.MatrixFloat()
    B = libfetchcore.fetch.math.linalg.MatrixFloat()
    C = libfetchcore.fetch.math.linalg.MatrixFloat()
    A.FromNumpy( np.random.rand(n,n) )
    B.FromNumpy( np.random.rand(n,n) )
    return B
def testGenNP(n):
    A = 100*np.random.rand(n,n)
    B = 100*np.random.rand(n,n)
    print A.dtype
    return B
def testDot(n):
    A = libfetchcore.fetch.math.linalg.MatrixFloat()
    B = libfetchcore.fetch.math.linalg.MatrixFloat()
    C = libfetchcore.fetch.math.linalg.MatrixFloat()
    A.FromNumpy( np.random.rand(n,n) )
    B.FromNumpy( np.random.rand(n,n) )
    return C.Dot(A,B)
def testDotReference(n):
    A = libfetchcore.fetch.math.linalg.MatrixFloat()
    B = libfetchcore.fetch.math.linalg.MatrixFloat()
    C = libfetchcore.fetch.math.linalg.MatrixFloat()
    A.FromNumpy( np.random.rand(n,n) )
    B.FromNumpy( np.random.rand(n,n) )
    return C.DotReference(A,B)
def testNP(n):
    A = np.random.rand(n,n)
    B = np.random.rand(n,n)
    return A.dot(B)
```

```python
M = 1
n = 1000
ref1 = timeit.timeit(lambda:testGen(n), number = M)
ref2 = timeit.timeit(lambda:testGenNP(n), number = M)
t1 = timeit.timeit(lambda:testDot(n), number = M)
t2 = timeit.timeit(lambda:testDotReference(n), number = M)
t3 = timeit.timeit(lambda:testNP(n), number = M)
print (t1-ref1) / M, (t2-ref1) / M, (t3 -ref2) / M
```
